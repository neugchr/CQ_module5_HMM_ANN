```python
from random import randint
from random import choice

class Test:
    def __init__(self, t=10, runs=1000):
        self.maxtime = t
        self.maxruns = runs
        self.time = 0
        self.state = choice(['A', 'B', 'C'])
        self.hist_A = [0] * t
        self.hist_B = [0] * t
        self.hist_C = [0] * t
        

    def run(self):
        for rn in range(self.maxruns):
            for tm in range(self.maxtime):
                if tm == 0:
                    self.state = 'A'
                r = randint(1,100)
                if self.state=='A':
                    self.hist_A[tm] +=1
                    if (r <=25):
                        self.state = 'A'
                    else:
                        self.state = 'B'
                elif self.state=='B':
                    self.hist_B[tm] +=1
                    if (r <= 50):
                        self.state = 'A'
                    else:
                        self.state = 'C'
                else:
                    self.hist_C[tm] +=1
                    if (r <=25):
                        self.state = 'C'
                    else:
                        self.state = 'B'
        

    def prt(self):
        norm_A = [None] * len(self.hist_A)
        norm_B = [None] * len(self.hist_B)
        norm_C = [None] * len(self.hist_C)
        print('\t\t counts \t | \t probs' )
        for i in range(len(self.hist_A)):
            norm_A[i] = (self.hist_A[i]/(self.hist_A[i]+self.hist_B[i]+self.hist_C[i]) )
            norm_B[i] = (self.hist_B[i]/(self.hist_A[i]+self.hist_B[i]+self.hist_C[i]) )
            norm_C[i] = (self.hist_C[i]/(self.hist_A[i]+self.hist_B[i]+self.hist_C[i]) )
            

            print(f'{i}:\t', self.hist_A[i], '\t', self.hist_B[i], '\t', self.hist_C[i],
                  '\t | \t', norm_A[i], '\t', norm_B[i], '\t',norm_C[i])


        
    
t=Test(t=50, runs=100)

t.run()
t.prt()
```
